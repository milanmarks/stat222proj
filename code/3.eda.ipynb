```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pointbiserialr
import pandas as pd
import numpy as np
```

```python
# Load the dataset
data = pd.read_csv('/Users/yimeng_yin/Desktop/STAT222/data/train_test_set.csv')
df = data[data["train"] == 1]
```

## Histograms and boxplots

```python
# Select a subset of the numerical columns for visualization
exclude_var = ['cik', 'sic', 'sic_description', 'accession_num', 'primary_doc', 'filelink', 'filedate' , 'year', 'main_industry', 'item_1A', 'item_2', 'item_7', "train", "clean_items", "sentiment"]

df_exclude_var = df.drop(columns = exclude_var)

# Separate the dataframe into two based on bank_status
df_bankrupt = df_exclude_var[df_exclude_var['bank_status'] == 1]
df_non_bankrupt = df_exclude_var[df_exclude_var['bank_status'] == 0]
palette = ['lightblue', 'salmon']

for column in df_bankrupt.columns[1:]: # Exclude the bank_status column
    plt.subplots(figsize = (40, 6.5))

    plt.subplot(1,4,1)
    plt.hist(df_non_bankrupt[column], bins = 20, color = 'lightblue', alpha = 0.7, label='Non-Bankrupt')
    plt.yscale('log')
    plt.xlabel(column, fontsize = 15)
    plt.ylabel('Frequency', fontsize = 15)
    plt.title(f'{column} Distribution (Non-Bankrupt)', fontsize = 18)

    plt.subplot(1,4,2)
    plt.hist(df_bankrupt[column], bins = 20, color = 'salmon', alpha = 0.7, label = 'Bankrupt')
    plt.yscale('log')
    plt.xlabel(column, fontsize = 15)
    plt.ylabel('Frequency', fontsize = 15)
    plt.title(f'{column} Distribution (Bankrupt)', fontsize = 18)

    plt.subplot(1,4,3)
    plt.hist(df_non_bankrupt[column], bins = 20, alpha = 0.7, label = 'Non-Bankrupt', color = 'lightblue')
    plt.hist(df_bankrupt[column], bins = 5, alpha = 0.7, label = 'Bankrupt', color = 'salmon')
    plt.yscale('log')
    plt.xlabel(column, fontsize = 15)
    plt.ylabel('Frequency', fontsize = 15)
    plt.title(f'{column} Distribution by Bankrupt Status', fontsize = 18)
    plt.legend(fontsize = 15)

    plt.subplot(1,4,4)
    sns.boxplot(x='bank_status', y = column, data = df, palette = palette, flierprops=dict(marker='o', markersize=1))
    for patch in plt.gca().artists:
        r, g, b, a = patch.get_facecolor()
        patch.set_facecolor((r, g, b, 0.1))
    plt.xlabel("Bankrupt status", fontsize = 15)
    plt.ylabel(column, fontsize = 15)
    plt.title(f'Boxplot of {column} by Bankrupt Status', fontsize = 18)
    plt.yscale('log')
    plt.tight_layout()

    plt.show()
```

## Bankruptcy rate across industries

```python
bankruptcy_counts = df.groupby('main_industry')['bank_status'].value_counts().unstack(fill_value=0)
data2 = bankruptcy_counts[1]
ratio = data2 / np.sum(data2)
ratio.plot(kind = "bar" , figsize = (14, 6) , ylabel = "Bankruptcy rate")
plt.legend(["# Bankruptcy within the industry / # Total bankruptcy"] , loc = "best")
plt.xlabel("Industry")
plt.xticks(rotation = 60)
plt.title("Bankruptcy Rate across Industries", fontsize = 20)

new_xtick_labels = ratio.index.tolist()
new_xtick_labels[-2] = "Transportation, Communications, etc."
plt.xticks(range(len(new_xtick_labels)), new_xtick_labels)

plt.show()
```

## Bankruptcy rate across year

```python
bankruptcy_counts = df.groupby('year')['bank_status'].value_counts().unstack(fill_value = 0)
data2 = bankruptcy_counts[1]
year_ratio = data2 / np.sum(bankruptcy_counts[1])
year_ratio.plot(kind = "bar" , figsize = (14, 6) , ylabel = "Bankruptcy rate")
plt.legend(["# Bankruptcy within the year / # Total bankruptcy"] , loc = "best")
plt.title("Bankruptcy Rate across Years", fontsize = 20)
plt.xticks(rotation = 0)
plt.xlabel("Year")
plt.show()
```

## Sentiment Score

```python
df_bankrupt = data[data['bank_status'] == 1]
df_non_bankrupt = data[data['bank_status'] == 0]
palette = ['lightblue', 'salmon']

plt.figure(figsize=(25, 5))
plt.subplot(1,4,1)
plt.hist(df_non_bankrupt['sentiment'], bins = 20, color = 'lightblue', alpha = 0.7, label='Non-Bankrupt')
plt.yscale('log')
plt.xlabel('Sentiment score')
plt.ylabel('Frequency')
plt.title('Sentiment Score Distribution (Non-Bankrupt)')

plt.subplot(1,4,2)
plt.hist(df_bankrupt['sentiment'], bins = 20, color = 'salmon', alpha = 0.7, label = 'Bankrupt')
plt.yscale('log')
plt.xlabel('Sentiment score')
plt.ylabel('Frequency')
plt.title('Sentiment Score Distribution (Bankrupt)')

plt.subplot(1,4,3)
plt.hist(df_non_bankrupt['sentiment'], bins=20, alpha=0.7, label='Non-Bankrupt', color = 'lightblue')
plt.hist(df_bankrupt['sentiment'], bins = 20, alpha=0.7, label='Bankrupt', color = 'salmon')
plt.yscale('log')
plt.xlabel('Sentiment score')
plt.ylabel('Frequency')
plt.title('Sentiment Score Distribution by Bankrupt Status')
plt.legend()

plt.subplot(1, 4, 4)
ax = sns.boxplot(x='bank_status', y='sentiment', data=df, palette=palette, flierprops=dict(marker='o', markersize=2))
ax.set_title('Boxplot of Sentiment by Bankrupt Status')
mean_values = data.groupby('bank_status')['sentiment'].mean()
labels = ['Non-Bankrupt', 'Bankrupt']
for i in range(len(mean_values)):
    ax.text(i, 1.02
            , f'Average: {mean_values[i]:.2f}', color='red', ha='center')
plt.xlabel("Bankrupt status")
plt.tight_layout()
plt.show()
```

## Topic Distribution

```python
lsa_data = np.load('../res/lsa.npy')

significant_topics = range(5)
significant_data = lsa_data[:, significant_topics]

y_train = df["bank_status"]
bankrupt_data = significant_data[y_train == 1]
non_bankrupt_data = significant_data[y_train == 0]

plt.figure(figsize=(18, 4))
for i, topic in enumerate(significant_topics):
    plt.subplot(1, 5, i + 1)
    sns.boxplot(x=y_train, y=significant_data[:, i], palette=["lightblue", "salmon"], flierprops=dict(marker='o', markersize=1))
    plt.title(f'Topic {topic + 1}')
    plt.xlabel('Bankrupt status')

plt.tight_layout()
plt.show()
```
